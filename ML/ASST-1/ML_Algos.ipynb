#Logistic Regression model making process

import libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, confusion_matrix
```

Step-1: Make classification date

```python
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, n_classes=2, random_state=42)
```

```python
X,y
```

Step-2: Train the Logistic regrassion model

```python
clf = LogisticRegression()
clf.fit(X, y)
```

Step-3: Prediction

```python
y_pred = clf.predict(X)
```

Step-4:Evaluation

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(7,4)})
sns.heatmap(confusion_matrix(y, y_pred), annot=True, fmt='d')

```

```python
print("Accuracy:", accuracy_score(y, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```

#KNN model making process

Import libraries

```python
from sklearn.neighbors import KNeighborsClassifier
```

Step-1: Create Model

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

Step-2: Train model

```python
knn.fit(X, y)
```

Step-3: prediction

```python
y_pred = knn.predict(X)
```

Step-4: Evaluation

```python
print("Accuracy:", accuracy_score(y, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y, y_pred))
```
